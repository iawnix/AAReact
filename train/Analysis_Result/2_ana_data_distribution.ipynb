```python
seed = 1
test_size = 0.2
prefix = "/home/iaw/DATA2/AAReact/DataSet/AtropicAcid_416/4_gen_feature/rdkit_soap_xtb_3"
data_x, data_y, x_label, data_class, data_name = load_data(
    "{}_data_x.npy".format(prefix), 
    "{}_data_y.npy".format(prefix), 
    "{}_x_label.pkl".format(prefix), 
    "{}_data_class.pkl".format(prefix), 
    "{}_data_name.pkl".format(prefix)
)
raw_csv_data = pd.read_csv("/home/iaw/DATA2/AAReact/DataSet/AtropicAcid_416/3_opted_mol2/brief_data_2026_02_03.csv")
```

```python
# 划分数据集
X_train, X_test, y_train, y_test,  class_train, class_test, name_train, name_test = train_test_split(
    data_x,        
    data_y,
    data_class,
    data_name, 
    test_size=test_size,
    random_state=seed, 
)

# 获取数据集中的cat标签
train_cat = []
for _data_id in name_train:
    _cat_name = raw_csv_data[raw_csv_data["DATA_ID"] == _data_id]["CAT_NAME"].values[0]
    train_cat.append(_cat_name)
test_cat = []
for _data_id in name_test:
    _cat_name = raw_csv_data[raw_csv_data["DATA_ID"] == _data_id]["CAT_NAME"].values[0]
    test_cat.append(_cat_name)

# 验证test中是否有train中不存在的cat
train_cat_set = set(train_cat)
test_cat_set = set(test_cat)

if (cat_diff:=test_cat_set - train_cat_set):
    print("test_cat_set - train_cat_set: {}".format(cat_diff))
else:
    print("test_cat_set - train_cat_set: Null")
```
